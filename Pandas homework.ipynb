```python
import pandas as pd

# Load the dataset
file_path = r"C:\Users\booke\Downloads\Python-Module-main\First project\MENTAL HEALTH DATA.csv"
df = pd.read_csv(file_path)

# Basic inspection
print(df.head())
print(df.info())
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())

# Check unique values in categorical columns
for col in df.select_dtypes(include='object').columns:
    print(f"{col}: {df[col].unique()}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter for Male and Female
filtered_df = df[df['sex'].isin(['Male', 'Female'])]

# Summarize data: Count occurrences of each general_health category by gender
summary = (
    filtered_df.groupby(['general_health', 'sex'])
    .size()
    .reset_index(name='count')
)

# Sort by general_health to ensure correct ordering
summary = summary.sort_values(by='general_health')

# Create a line plot
sns.lineplot(data=summary, x='general_health', y='count', hue='sex', marker='o')

# Customize the plot
plt.title('General Health by Gender')
plt.xlabel('General Health Categories')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.legend(title='Gender')
plt.grid(True)
plt.show()
```

```python
# Summarize data: Count social media usage by age and sex
summary = df.groupby(['age', 'sex', 'social_media']).size().reset_index(name='count')

# Normalize counts to calculate percentages for social media usage
summary['percentage'] = (
    summary.groupby(['age', 'sex'])['count'].apply(lambda x: 100 * x / x.sum())
)
```
